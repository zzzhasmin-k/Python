# Методы класса

Методы класса представляют функции, которые определены внутри класса и которые определяют его поведение

```python
class Verb:

  @classmethod
  def check(cls, arg):
    if cls.MIN_LEN <= len(arg) <= cls.MAX_LEN and arg.isalpha():
      return True
    else:
      raise ValueError('Length out of range or Nonalphabetical symbols detected')

  MIN_LEN = 2
  MAX_LEN = 20

  def __init__(self, lemma='do', past='did', past_participle='done'):
    if Verb.check(lemma) and Verb.check(past) and Verb.check(past_participle):
      self.lemma = lemma
      self.past = past
      self.past_participle = past_participle

  def print_info(self):
    print(f"Lemma: {self.lemma}", f"Past form: {self.past}", f"Past participle: {self.past_participle}", sep='\n')
```

Запись @classmethod вызывает встроенный декоратор. Это означает, что можно использовать свойства класса внутри этого метода без создания конкретного экземпляра класса.

```python
g = Verb()
```

```python
g.print_info()
```

```python
g = Verb('123')
```

# Декораторы и работа с ними

По большому счёту, декоратор представляет собой дополнительную "надстройку к функции", которая модифицирует её поведение. В Python есть встроенные декораторы; писать свои имеет смысл тогда, когда нам по какой-либо причине надо модифицировать действие ряда функций одинаковым образом. Например, чтобы программа выводила на экран, какая функция в данный момент исполняется.

```python
def get_func_name(func):
  def wrapper(*args, **kwargs):
    res = func(*args, **kwargs)
    print(func.__name__) # здесь мы указываем, что именно хотим от декоратора
    return res
  return wrapper
```

```python
@get_func_name
def plus_two(n):
  return n + 2
```

2. Напишите итератор, который будет подавать в цикл переданное ему число элементов. Например, если мы итерируемся по списку [1, 2, 3, 4, 5], а на вход итератору передано число 2, он сначала вернёт (1, 2), затем (3, 4) и на этом завершит итерации.

```python
class BatchIterator:
    def __init__(self, lst, batch_size):
        self.lst = lst
        self.batch_size = batch_size
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.lst):
            raise StopIteration

        batch = self.lst[self.index : self.index + self.batch_size]
        self.index += self.batch_size
        return tuple(batch)


lst = [1, 2, 3, 4, 5]
for b in BatchIterator(lst, 2):
    print(b)
```

3. Напишите класс, на вход которому при создании экземпляра подаётся список строк, в каждой из которых обязательно содержится как минимум 2 слова. Класс должен поддерживать следующие методы:

- get_tf(word, number of doc): принимает на вход слово и номер строки, возвращает значение TF

- get_idf(word, number of doc): принимает на вход те же самые параметры, возвращает значение IDF

- get_tf_idf(word, number of doc, ignore_stopwords=True): возвращает значение TF-IDF. По умолчанию при его вычислении не учитываются стопслова, если при вызове данного метода установить значение ignore_stopwords=False, учитывать стопслова при вычислении TF-IDF. Стопслова сделайте переменной класса.

```python
import math

class TextTFIDF:
    stopwords = {"and", "the", "a", "of", "to", "in", "on"}

    def __init__(self, docs):
        self.docs = [d.lower().split() for d in docs]

    def get_tf(self, word, doc_id):
        word = word.lower()
        doc = self.docs[doc_id]

        return doc.count(word) / len(doc)

    def get_idf(self, word):
        word = word.lower()
        N = len(self.docs)

        df = sum(1 for doc in self.docs if word in doc)

        if df == 0:
            return 0

        return math.log(N / df)

    def get_tf_idf(self, word, doc_id, ignore_stopwords=True):
        word = word.lower()

        if ignore_stopwords and word in TextTFIDF.stopwords:
            return 0

        tf = self.get_tf(word, doc_id)
        idf = self.get_idf(word)

        return tf * idf
```

```python
docs = [
    "I love natural language processing",
    "Language models love data",
    "Processing language is fun"
]

tfidf = TextTFIDF(docs)

print("TF:", tfidf.get_tf("language", 0))
print("IDF:", tfidf.get_idf("language"))
print("TF-IDF:", tfidf.get_tf_idf("language", 0))
```
