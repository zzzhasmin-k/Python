# Практикум

Написать программу, которая получает на вход три числа и выводит квадрат наибольшего из них, если оно больше 0.

```python
x = int(input(("Число 1:")))
y = int(input(("Число 2:")))
z = int(input(("Число 3:")))
max_num = max(x,y,z)
if max_num > 0:
  print(max_num**2)
```

Написать программу, которая получает на вход три числа и вычисляет площадь треугольника согласно формуле Герона (её можно найти в интернете).

```python
x = int(input(("Число 1:")))
y = int(input(("Число 2:")))
z = int(input(("Число 3:")))
triangle = ( x+ y + z) / 2
print(triangle)
```

Написать программу, которая получает на вход три числа и выводит их сначала максимальное, потом минимальное, а потом оставшееся. Предусмотреть случай, когда два числа равны друг другу!

```python
x = int(input(("Число 1:")))
y = int(input(("Число 2:")))
z = int(input(("Число 3:")))
max_num = max(x, y, z)
min_num = min(x, y, z)
remaining = (x + y + z) - max_num - min_num
print(max_num)
print(min_num)
print(remaining)
```

Написать программу, которая вычисляет процентное соотношение гласных букв в слове вне зависимости от регистра. Слово подается на вход стандартным вводом.

```python
word = input().strip().lower()
vowels = "aeiou"
vowels_only = word
for consonant in "bcdfghjklmnpqrstvwxyz":
    vowels_only = vowels_only.replace(consonant, "")

if len(word) > 0:
      percentage = (len(vowels_only) / len(word)) * 100
      print(round(percentage, 2))
else:
      print("0.00")
```

Написать программу, которая проверяет, является ли введенная строка палиндромом. Регистр и знаки препинания (для простоты эксперимента - только точки и запятые) не учитываются.

```python
line = input()
cleaned_line= line.lower().replace('.', '').replace(',', '')
if cleaned_line == cleaned_line[::-1]:
    print("True")
else:
    print("False")
```
